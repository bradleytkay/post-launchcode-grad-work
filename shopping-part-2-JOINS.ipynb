## **Let's practice JOINS**

### **1\. Use the foreign key customer\_id, which connects the customers, orders, and reviews tables**

```sql
SELECT TOP 1 *
FROM OnlineShoppingDB.dbo.customers
```

```sql
SELECT TOP 1 *
FROM OnlineShoppingDB.dbo.orders
```

```sql
SELECT TOP 1 *
FROM OnlineShoppingDB.dbo.reviews
```

```sql
SELECT 
    c.first_name, 
    c.last_name, 
    o.order_date, 
    o.total_price, 
    r.rating, 
    r.review_text
FROM 
    OnlineShoppingDB.dbo.customers AS c
INNER JOIN
    OnlineShoppingDB.dbo.orders AS o 
ON 
    c.customer_id = o.customer_id 
INNER JOIN
    OnlineShoppingDB.dbo.reviews AS r 
ON
    o.customer_id = r.customer_id

--Just joined more than 2 tables for the first time.  I know this isn't a big deal, but this isn't
--something we learned how to do at LaunchCode, so I feel a tiny bit accomplished :) 

--Why am I seeing duplicate rows, though?
```
