### **1\. Let's practice some UNIONs.**

**A. List all customers who have placed orders or reviewed items.**

```sql
SELECT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT DISTINCT customer_id
    FROM OnlineShoppingDB.dbo.orders 
)
```

```sql
SELECT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT DISTINCT customer_id
    FROM OnlineShoppingDB.dbo.reviews
)
```

```sql
SELECT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT DISTINCT customer_id
    FROM OnlineShoppingDB.dbo.orders 
)
UNION 
SELECT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT DISTINCT customer_id
    FROM OnlineShoppingDB.dbo.reviews 
)
```

**B. List all products from 2 different categories.**

```sql
SELECT DISTINCT category
FROM OnlineShoppingDB.dbo.products 
```

```sql
SELECT product_name
FROM OnlineShoppingDB.dbo.products 
WHERE category = 'Electronics'
UNION
SELECT product_name
FROM OnlineShoppingDB.dbo.products 
WHERE category = 'Furniture'
```

### **2\. Now for some INTERSECTs.**

A. Find customers who ordered a product AND left a review.

```sql
SELECT DISTINCT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT customer_id
    FROM OnlineShoppingDB.dbo.orders 
    INTERSECT
    SELECT customer_id
    FROM OnlineShoppingDB.dbo.reviews 
)

--OR

-- SELECT first_name, last_name
-- FROM OnlineShoppingDB.dbo.customers
-- WHERE customer_id IN (
--     SELECT DISTINCT customer_id
--     FROM OnlineShoppingDB.dbo.orders 
-- )
-- UNION 
-- SELECT first_name, last_name
-- FROM OnlineShoppingDB.dbo.customers
-- WHERE customer_id IN (
--     SELECT DISTINCT customer_id
--     FROM OnlineShoppingDB.dbo.reviews 
-- )
```

**B. Find products that were ordered AND reviewed.**

```sql
SELECT DISTINCT product_name
FROM OnlineShoppingDB.dbo.products
WHERE product_id IN (
    SELECT product_id
    FROM OnlineShoppingDB.dbo.order_items
    INTERSECT
    SELECT product_id
    FROM OnlineShoppingDB.dbo.reviews 
)
```

**C. Find customers who wrote reviews BEFORE receiving shipments.**

```sql
SELECT r.customer_id
FROM OnlineShoppingDB.dbo.reviews AS r 
WHERE r.review_date < (
    SELECT TOP 1 s.delivery_date
    FROM OnlineShoppingDB.dbo.shipments AS s
    WHERE s.order_id IN (
        SELECT oi.order_id
        FROM OnlineShoppingDB.dbo.order_items AS oi 
        WHERE oi.product_id = r.product_id
        AND oi.order_id = s.order_id
    )
    ORDER BY s.delivery_date
)
INTERSECT
SELECT o.customer_id 
FROM OnlineShoppingDB.dbo.orders AS o
```

### **3\. Finally, on to EXCEPTs.**

**A. Find customers who ordered an item but DIDN'T leave a review.**

```sql
SELECT DISTINCT first_name, last_name
FROM OnlineShoppingDB.dbo.customers
WHERE customer_id IN (
    SELECT customer_id
    FROM OnlineShoppingDB.dbo.orders 
    EXCEPT
    SELECT customer_id
    FROM OnlineShoppingDB.dbo.reviews 
)
```
